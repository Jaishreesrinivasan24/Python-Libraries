```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
cal= fetch_california_housing()
```

```python
cal
```

```python
print(cal.DESCR)
```

```python
cal.keys()
```

```python
cal.data
```

```python
cal.feature_names
```

```python
df=pd.DataFrame(cal.data,columns=cal.feature_names)
```

```python
df
```

```python
cal.target
```

```python
df['Price']=cal.target
```

```python
df
```

```python
#descriptive analytics
df.describe()
```

```python
#check null values
df.info()
```

```python
#another way to check null values
df.isnull().sum()
```

```python
#drops the record which has null value
df.dropna()
```

```python
#data is now cleaned, next step is Data exploration
#used scatterplot as bivariate and multivariate visualisation
sns.scatterplot(x='Longitude',y='Latitude',data=df, hue= 'Price', size= 'AveOccup')
```

```python
#gives multiple graphs is pairplot
sns.pairplot(df[['Price','AveBedrms','MedInc']])
```

```python
#histogram plot
sns.histplot(df['Price'], bins=30,kde=True)
plt.savefig('Histogram for price.png')
```

```python
#increase size of the image
plt.figure(figsize=(15,6))
#heatmap
sns.heatmap(data=df.corr(),annot=True, cmap='coolwarm')
```

```python
X=df.drop('Price',axis =1)
X
```

```python
Y=df['Price']
Y
```

```python
#model training
#step 1:Divide the dataset into two parts
from sklearn.model_selection import train_test_split
#above converts method into two parts
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, random_state=42, test_size= .30)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Reg=LinearRegression()
```

```python
Reg.fit(X_train, Y_train)
```

```python
#predicted
y_pred=Reg.predict(X_test)
```

```python
#r2_score checks the accurate value, mean_sq_error compares the actual minus predicted
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_pred, Y_test)
```
